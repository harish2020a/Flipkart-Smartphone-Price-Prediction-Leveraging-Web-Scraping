**WEB SCRAPER**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}
```

```python
base_url = 'https://www.flipkart.com/search?q=smart+phones&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_5_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_5_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=smart+phones&requestId=bfc61a51-3e10-4f01-99d2-5bbf7ac9e5ef&as-searchtext=smart&page='
```

```python
pages = 41
```

```python
df = pd.DataFrame()
```

```python
for page in range(1,pages+1):
    r = requests.get(base_url + str(page), headers=header)
    soup = BeautifulSoup(r.content, 'html.parser')
    mobiles = soup.find_all('a', class_='_1fQZEK')
    for mobile in mobiles:
      try:
        row = {}
        row['name'] = mobile.find('div', class_='_4rR01T').text.strip(
        ) if mobile.find('div', class_='_4rR01T') else float('nan')
        features = mobile.find_all('li', class_='rgWa7D')
        row['memory'] = features[0].text.strip() if features[0] else float('nan')
        row['display'] = features[1].text.strip() if features[1] else float('nan')
        row['camera'] = features[2].text.strip() if features[2] else float('nan')
        row['power'] = features[3].text.strip() if features[3] else float('nan')
        row['price'] = mobile.find('div', class_='_30jeq3 _1_WHN1').text.strip(
        ) if mobile.find('div', class_='_30jeq3 _1_WHN1') else float('nan')
        row['discount'] = mobile.find('div', class_='_3Ay6Sb').find('span').text.strip(
        ) if mobile.find('div', class_='_3Ay6Sb').find('span') else float('nan')
        row['rating'] = mobile.find('div', class_='_3LWZlK').text.strip(
        ) if mobile.find('div', class_='_3LWZlK') else float('nan')
        df = df.append(row, ignore_index=True)
      except:
        print('Exception')
```

**Machine Learning Models Implementation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('flipkart mobiles.csv')
```

```python
df.head()
```

```python
df[["RAM", "ROM", "Expandable"]] = df["memory"].str.split("|", expand=True)
df.drop("memory", axis=1, inplace=True)
df['display'] = (df["display"].str.split(" cm", expand=True)[0]).astype(float)
df['power'] = (df["power"].str.split(" mAh", expand=True)[0])
df['power'] = df['power'].apply(lambda x: int(x) if x.isnumeric() else None)
df['price'] = (df['price'].str.replace('₹', '')).str.replace(',', '').astype('int64')
df['discount'] = df['discount'].str.replace('% off', '').astype('int64')
df['RAM'] = (df["RAM"].str.split(" ", expand=True)[0]).astype('int64')
df['ROM'] = (df["ROM"].str.split(" ", expand=True)[1])
df['ROM'] = df['ROM'].apply(lambda x: None if x is None else int(x))
df['Expandable'] = (((df['Expandable'].str.replace(' Expandable Upto ', '')).str.replace(' TB', '000000')).str.replace(' GB','000')).str.replace(' MB', '')
df['Expandable'] = df['Expandable'].apply(lambda x: None if x is None else int(x))
df['camera'] = df["camera"].str[:2]
df['camera'] = df['camera'].apply(lambda x: int(x) if x.isnumeric() else None)
```

```python
df.isnull().any()
```

```python
df['camera'].fillna(value=df['camera'].mean(), inplace=True)
df['power'].fillna(value=df['power'].mean(), inplace=True)
df['rating'].fillna(value=df['rating'].mean(), inplace=True)
df['ROM'].fillna(value=df['ROM'].mean(), inplace=True)
df['Expandable'].fillna(value=df['Expandable'].mean(), inplace=True)
```

```python
df.head()
```

```python
df.shape
```

**Random Forest**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(columns=['price', 'name'])
y = np.log(df['price'])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Regression Report:")
print("-------------------")
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
